```python
# imports

import os
from dotenv import load_dotenv
from IPython.display import Markdown, display
from openai import OpenAI

# If you get an error running this cell, then please head over to the troubleshooting notebook!
```

```python
load_dotenv(override=True)  # Load environment variables from .env file
api_key = os.getenv("OPENAI_API_KEY")  # Get the API key from environment variables

if not api_key:
    raise ValueError("OPENAI_API_KEY not found in environment variables. Please set it in your .env file.") 
elif api_key.startswith("sk-"):
    print("API key loaded successfully.")
```

Movie Reviewer Platform, which will tell about any movie

```python
openai = OpenAI()
```

```python
# Step 1: Create your prompts

system_prompt = """You are a movie expert that provides crisp and easy to understand summaries of movies.
You are very concise and only include the most important details about the plot, characters, and themes of the movie. 
You do not include any spoilers in your summaries. Your response is very polite and friendly as if someone asking you as a friend for a recommendation."""
user_prompt = """
    I am thinking of watching a movie which just hit in India Bollywood and became a blockbuster.
    Can you tell me a bit about Dhurandhar - The revenge, which just has become a blockbuster?
"""

# Step 2: Make the messages list

messages = [
    {"role": "system", "content": system_prompt},
    {"role": "user", "content": user_prompt}
]

# Step 3: Call OpenAI
response = openai.chat.completions.create(model="gpt-4.1-mini", messages=messages)


# Step 4: print the result
print(response.choices[0].message.content)
```

```python
def summarize_movie(movie_name):
    system_prompt = """You are a movie expert that provides crisp and easy to understand summaries of movies.
    You are very concise and only include the most important details about the plot, characters, and themes of the movie. 
    You do not include any spoilers in your summaries. 
    Your response is rude and harsh tone  as if someone asking you as a friend but multiple times in a day."""
    user_prompt = f"""
        I am thinking of watching a movie which just hit in India Bollywood and became a blockbuster.
        Can you tell me a bit about {movie_name}, which just has become a blockbuster?
    """
    messages = [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt}
    ]
    response = openai.chat.completions.create(model="gpt-4.1-mini", messages=messages)
    return response.choices[0].message.content

def display_movie_summary(movie_name):
    summary = summarize_movie(movie_name)
    display(Markdown(summary))

display_movie_summary("Dhurandhar - The revenge")
print("\n")
display_movie_summary("Koi mil gya")
```
