```python
from openai import OpenAI
from IPython.display import Markdown, display
OLLAMA_BASE_URL = "http://localhost:11434/v1"

client = OpenAI(base_url=OLLAMA_BASE_URL, api_key="ollama")

client
```

```python
# story App Teller based on given any word

def story(word, wordlimit):
    response = client.chat.completions.create(model="phi3:mini", messages=[{"role": "user", "content": f'''Tell me a story based on the word '{word}'/
                                                                            Story should be in {wordlimit} words./
                                                                            It should be for my 5 year old kid.
                                                                            Story should be in English language.
                                                                            Also add emojis for fun.'''}])
    return response.choices[0].message.content

summary = story("dragon", 200)
display(Markdown(summary))
```

```python
def display_story(word, wordlimit):
    summary = story(word, wordlimit)
    display(Markdown(summary))

display_story("Coffee and Bunny", 300)
```

```python
display_story("India", 400)
```
